# ***E-Commerce (EDA)***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px  # for Visualization
import plotly.graph_objects as go # for customized & advanced graphs
import plotly.io as pio # for graph template customization
import plotly.colors as colors
pio.templates.default = "plotly_white"
```

```python
df = pd.read_csv("Sample - Superstore.csv", encoding="latin-1")
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.info()
```

Converting date columns

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.info()
```

```python
df['Order_Month']=df['Order Date'].dt.month
df['Order_Year']=df['Order Date'].dt.year
df['Order_Day_of _weak']=df['Order Date'].dt.dayofweek
```

```python
df.head()
```

1. Calculate Monthly Sales of the store and Identify which month had highest sales an which month had the lowest sales

```python
sales_by_month = df.groupby('Order_Month')['Sales'].sum().reset_index()
sales_by_month
```

```python
fig = px.line(sales_by_month, x='Order_Month', y='Sales', title='Monthly Sales of the Store')
fig.show()
```

2. Analyze sales based on product categories and determinewhich category has the lowest sales and which category has the highest sales.

```python
sales_by_category = df.groupby('Category')['Sales'].sum().reset_index()
sales_by_category
```

```python
fig = px.pie(sales_by_category, values='Sales', names='Category',hole = 0.5,
             color_discrete_sequence= px.colors.qualitative.Pastel)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Sales by Category', title_font=dict(size=24))

fig.show()
```

3. Sales Analysis based on sub-categories

```python
sales_by_subcategory = df.groupby(['Category', 'Sub-Category'])['Sales'].sum().reset_index()
sales_by_subcategory
```

```python
fig = px.bar(sales_by_subcategory, x='Sub-Category', y='Sales', color='Category', barmode='group',
             title='Sales Analysis by Sub-Category')
fig.show()
```

4. Analyze the monthly profit from sales and determine which month had the highest profit.

```python
profit_by_month = df.groupby('Order_Month')['Profit'].sum().reset_index()
profit_by_month
```

```python
fig = px.bar(profit_by_month, x='Order_Month', y='Profit', title='Monthly Profit from Sales')
fig.show()
```

5. analyze profit by category and Sub-Category

```python
profit_by_category = df.groupby('Category')['Profit'].sum().reset_index()
profit_by_category
```

```python
fig = px.pie(profit_by_category, values='Profit', names='Category', hole=0.5,
             color_discrete_sequence=px.colors.qualitative.Pastel)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Profit by Category', title_font=dict(size=24))

fig.show()
```

```python
profit_by_subcategory = df.groupby('Sub-Category')['Profit'].sum().reset_index()
profit_by_subcategory
```

```python
fig = px.bar(profit_by_subcategory, x='Sub-Category', y='Profit', title='Profit by Sub-Category')
fig.show()
```

6. Analyze sales and Profit by customer segment

```python
s_p_by_segment = df.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
```

```python
color_palette = colors.qualitative.Pastel

fig = go.Figure()
fig.add_trace(go.Bar(x=s_p_by_segment['Segment'],
                     y=s_p_by_segment['Sales'],
                     name='Sales',
                     marker_color=color_palette[0]))

fig.add_trace(go.Bar(x=s_p_by_segment['Segment'],
                     y=s_p_by_segment['Profit'],
                     name='Profit',
                     marker_color=color_palette[1]))

fig.update_layout(title_text='Sales and Profit by Customer Segment',
                  xaxis_title='Customer Segment',
                  yaxis_title='Amount')

fig.show()
```

7. Analyze sales to profit ratio

```python
sales_profit_segment = df.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
sales_profit_segment['Profit_Ratio'] = sales_profit_segment['Sales'] / sales_profit_segment['Profit']
print(sales_profit_segment[['Segment','Profit_Ratio']])
```

# ***Conclusion :***
The EDA revealed that sales peaked in November and were lowest in February,
while profits were highest in December and lowest in January.
Technology emerged as the most profitable category, with Phones and Copiers
leading at the sub-category level, whereas Office Supplies underperformed.
Among customer segments, Consumers contributed the highest sales and profit ratio,while the Home Office segment had the lowest sales and profitability.
